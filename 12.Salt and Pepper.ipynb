```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('photo.PNG')
```

```python
plt.figure(figsize = (12,6))
plt.imshow(img)
plt.axis('off')
```

```python
def add_noise(image, salt_prob, pepper_prob):
    noisy_image = np.copy(image)

    # Salt noise 
    salt_noise = np.random.rand(*image.shape) < salt_prob
    noisy_image[salt_noise] = 255

    # Pepper noise 
    pepper_noise = np.random.rand(*image.shape) < pepper_prob
    noisy_image[pepper_noise] = 0

    return noisy_image
```

```python
img = cv2.imread('photo.PNG', 0) 
salt_probs = [0.05, 0.50, 0.95]
pepper_probs = [1 - prob for prob in salt_probs]
```

```python
plt.figure(figsize=(18, 6))

for i, salt_prob in enumerate(salt_probs, 1):
    noisy_img = add_noise(img, salt_prob, pepper_probs[i-1])

    plt.subplot(1, len(salt_probs), i)
    plt.title(f'Salt Probability: {salt_prob}\nPepper Probability: {pepper_probs[i-1]:.2f}')
    plt.imshow(noisy_img, cmap='gray')
    plt.axis('off')

plt.show()
```
