```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve


def sobel_filter(image):
    sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

    gradient_x = convolve(image, sobel_x)
    gradient_y = convolve(image, sobel_y)

    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
    gradient_angle = np.arctan2(gradient_y, gradient_x)

    return gradient_magnitude, gradient_angle
# rnek bir gri tonlu grnt
image = plt.imread('photo.PNG')[:,:,0]

# Sobel filtresini uygula
gradient_magnitude, gradient_angle = sobel_filter(image)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Sobel Filtered Image')
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Sobel Gradient Angle')
plt.imshow(gradient_angle, cmap='hsv')
plt.axis('off')

plt.show()
```
