```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams["figure.autolayout"] = True
```

```python
img = cv2.imread('photo.PNG')
```

```python
plt.figure(figsize = (12,6))
plt.imshow(img)
plt.axis('off')
```

```python
def gaussianFilter(img, kernel_size, sigma):
    kernel = cv2.getGaussianKernel(kernel_size, sigma)
    kernel_2d = np.outer(kernel, kernel.T)
    filtered_img = cv2.filter2D(img, -1, kernel_2d)
    return filtered_img
```

```python
kernel_size = 3
sigma_values = [0.5, 1.0 , 2.0] 
smoothed_img = gaussianFilter(img, kernel_size, sigma)
```

```python
plt.figure(figsize=(18, 6))

for i, sigma in enumerate(sigma_values, 1):
    smoothed_img = gaussianFilter(img, kernel_size, sigma)

    plt.subplot(1, len(sigma_values), i)
    plt.title(f'Sigma: {sigma}')
    plt.imshow(smoothed_img, cmap='gray')
    plt.axis('off')

plt.show()
```
