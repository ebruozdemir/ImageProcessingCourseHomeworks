# Normalization

The normalization process is scaling image pixels between 0 and 1.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def normalize_image(image):
    # Normalizasyon
    normalized_image = image / 255.0
    return normalized_image

img = cv2.imread('photo.PNG', cv2.IMREAD_GRAYSCALE)

normalized_img = normalize_image(img)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Normalized Image')
plt.imshow(normalized_img, cmap='gray')
plt.axis('off')
```

# Standardization

Standardization is the transformation of image pixels to have a mean of 0 and a standard deviation of 1. The standardization process tries to make the data resemble a normal distribution.

```python
def standardize_image(image):
    # Standardizasyon
    mean = np.mean(image)
    std = np.std(image)
    standardized_image = (image - mean) / std
    return standardized_image

img = cv2.imread('photo.PNG', cv2.IMREAD_GRAYSCALE)

standardized_img = standardize_image(img)


plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Standardizated Image')
plt.imshow(standardized_img, cmap='gray')
plt.axis('off')

plt.show()
```

# Differences Between Normalization and Standardization

-Normalization scales values within a given range, standardization converts values to a given mean and standard deviation.

-Normalization normalizes each value in the data set by dividing it by the difference between the minimum and maximum values of that value, while standardization normalizes each value in the data set by subtracting it from the mean of that value and dividing it by the standard deviation.

-Normalization is especially widely used for algorithms such as deep learning and neural networks, while standardization is generally preferred for algorithms that work with statistical methods based on the assumption of normal distribution.
