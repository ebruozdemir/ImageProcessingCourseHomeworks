```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams["figure.autolayout"] = True
```

```python
img = cv2.imread('histogram.PNG', 0)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
equ = cv2.equalizeHist(img)
```

```python
plt.figure(figsize=(18, 6))

# Original Histogram
plt.subplot(1, 3, 1)
plt.hist(img.flatten(), 256, [0, 256], color='r', alpha=0.5, label='Original')
plt.legend(loc='upper right')

# Equalized Histogram
plt.subplot(1, 3, 2)
plt.hist(equ.flatten(), 256, [0, 256], color='b', alpha=0.5, label='Equalized')
plt.legend(loc='upper right')

# Original and Equalized Histograms in the same plot
plt.subplot(1, 3, 3)
plt.hist(img.flatten(), 256, [0, 256], color='r', alpha=0.5, label='Original')
plt.hist(equ.flatten(), 256, [0, 256], color='b', alpha=0.5, label='Equalized')
plt.legend(loc='upper right')

plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Orginal Image')
plt.imshow(img, cmap='gray')
plt.axis('off')
    

plt.subplot(1, 2, 2)
plt.title('Histogram Equalized Image')
plt.imshow(equ, cmap='gray')
plt.axis('off')
```
