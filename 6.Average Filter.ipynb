```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams["figure.autolayout"] = True
```

```python
img = cv2.imread('mean.PNG',0)
```

```python
def average_filter(img, kernel_size):
    kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)
    filtered_img = cv2.filter2D(img, -1, kernel)
    return filtered_img
```

```python
kernel_sizes = [5, 7, 9]

plt.figure(figsize=(12, 6 * len(kernel_sizes)))

for i, kernel_size in enumerate(kernel_sizes, 1):
    smoothed_img = average_filter(img, kernel_size)

    plt.subplot(len(kernel_sizes), 2, 2 * i - 1)
    plt.title(f'Original Image\n')
    plt.imshow(img, cmap='gray')
    plt.axis('off')

    plt.subplot(len(kernel_sizes), 2, 2 * i)
    plt.title(f'Average Filtered Image\nKernel Size: {kernel_size}')
    plt.imshow(smoothed_img, cmap='gray')
    plt.axis('off')

plt.show()
```
