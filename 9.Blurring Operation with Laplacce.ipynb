```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def laplacian_blur(image):
    laplace_kernel = np.array([[0, 1, 0],
                               [1, -4, 1],
                               [0, 1, 0]])

    laplace_filtered_img = cv2.filter2D(image, cv2.CV_64F, laplace_kernel)

    # Normalized Process
    laplace_filtered_img = cv2.normalize(laplace_filtered_img, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)

    return laplace_filtered_img



image = cv2.imread('photo.PNG', cv2.IMREAD_GRAYSCALE)

laplacian_result = laplacian_blur(image)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Laplacian Blurred Image')
plt.imshow(laplacian_result, cmap='gray')
plt.axis('off')

plt.show()
```
