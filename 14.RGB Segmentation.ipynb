```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def rgb_thresholding(image, threshold):
    #If pixel greater than threshold ,make white ; otherwise black
    thresholded_image = np.zeros_like(image)
    thresholded_image[image > threshold] = 255

    return thresholded_image

img = cv2.imread('photo.PNG')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

threshold_value = 100

thresholded_img = rgb_thresholding(img_rgb, threshold_value)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Orjinal Resim')
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('RGB Bölütlenmiş Resim')
plt.imshow(thresholded_img, cmap='gray')
plt.axis('off')

plt.show()
```
