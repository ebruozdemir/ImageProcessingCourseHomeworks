```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('mean.PNG')
rows, cols = img.shape[:2]

kernel_size = (5, 5)

# Kernel Blurring using filter2D()
kernel_25 = np.ones((25, 25), np.float32) / 625.0
output_kernel = cv2.filter2D(img, -1, kernel_25)

# Boxfilter 
output_box = cv2.boxFilter(img, -1, kernel_size, normalize=True)
output_box = cv2.cvtColor(output_box, cv2.COLOR_BGR2RGB)  # Convert to RGB

# Gaussian Blur
output_gaus = cv2.GaussianBlur(img, kernel_size, 0)

# Median Blur (reduction of noise)
output_med = cv2.medianBlur(img, 5)

# Bilateral filtering (Reduction of noise + Preserving of edges)
output_bil = cv2.bilateralFilter(img, 5, 6, 6)

# Sharpening filter
sharpening_filter = np.array([[-1, -1, -1],
                              [-1, 9, -1],
                              [-1, -1, -1]])
output_sharpen = cv2.filter2D(img, -1, sharpening_filter)


plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(2, 3, 2), plt.imshow(cv2.cvtColor(output_gaus, cv2.COLOR_BGR2RGB)), plt.title('Gaussian Blur')
plt.subplot(2, 3, 3), plt.imshow(output_box), plt.title('Box Filter')
plt.subplot(2, 3, 4), plt.imshow(cv2.cvtColor(output_kernel, cv2.COLOR_BGR2RGB)), plt.title('Kernel Blur')
plt.subplot(2, 3, 5), plt.imshow(cv2.cvtColor(output_med, cv2.COLOR_BGR2RGB)), plt.title('Median Blur')
plt.subplot(2, 3, 6), plt.imshow(cv2.cvtColor(output_bil, cv2.COLOR_BGR2RGB)), plt.title('Bilateral Blur')


plt.figure(figsize=(12,8))
plt.subplot(1, 1, 1), plt.imshow(cv2.cvtColor(output_sharpen, cv2.COLOR_BGR2RGB)), plt.title('Sharpening Filter')


plt.show()
```
